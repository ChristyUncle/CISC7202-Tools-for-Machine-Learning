```python
import numpy as np
import sklearn
import matplotlib.pyplot as plt
```

```python
#1 
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()
# print('Keys of breast cancer: \n{}'.format(cancer.keys()))
# print("shape of cancer data:",cancer['data'].shape)
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify = cancer.target, random_state = 1)
```

```python
#2 Curve for training accuracy and test accuracy when K increases from 1 to 30
from sklearn.neighbors import KNeighborsClassifier

training_accuracy = []
test_accuracy = []

neighbors_setting = range(1, 31)

for n_neighbors in neighbors_setting:
    clf = KNeighborsClassifier(n_neighbors = n_neighbors)
    clf.fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))

plt.plot(neighbors_setting, training_accuracy, label = "training accuracy")
plt.plot(neighbors_setting, test_accuracy, label = 'test accuracy')
plt.ylabel("Accuracy")
plt.xlabel("N Neighbors")
plt.legend()
```
